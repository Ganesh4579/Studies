```python
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
print(df.head())
print(df.tail())
print(df.columns)
```

```python
x=df.iloc[:,:1]
y=df.iloc[:,1:]/1000
```

```python
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(x,y)#,test_size=.2)
```

```python
def lr(a,b):
     from sklearn.linear_model import LinearRegression
     reg=LinearRegression()
     reg.fit(a,b)
     return reg
```

```python
reg=lr(xtr,ytr)
print('m :',reg.coef_)
print('m :',reg.intercept_)
```

```python
print('train score :',reg.score(xtr,ytr))
print('test score :',reg.score(xte,yte))
```

```python
import matplotlib.pyplot as plt
plt.scatter(xtr,ytr)
plt.plot(xte,reg.predict(xte))
plt.show()
```
