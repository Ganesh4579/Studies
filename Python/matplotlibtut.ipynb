```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,20)
y=np.linspace(100,200,20)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r')
plt.subplot(2,1,2)
plt.plot(x,y,'b')

```

```python
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.9,0.9])
ax1.plot(x,y,label='first')
ax1.plot(y,x**2,label='first')
ax1.legend(loc=0)

ax2=fig.add_axes([0.45,0.45,0.4,0.3])
ax2.plot(x,y,label='first')
ax2.plot(y,x**2,label='first')
ax2.legend(loc=0)
ax2.text(0,40,'HEY')
ax2.grid(True)
plt.show()
```
